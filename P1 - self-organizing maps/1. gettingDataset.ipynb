```python
from pattern.web import Twitter, plaintext
from pattern.en import tag
import re

twitter = Twitter()
file = open('datasetG.txt', 'w+')

for tweet in twitter.search("Machine Learning", cached=False, count=100):
    s = plaintext(tweet.text.lower())
    print("Tweet: " + s + "\n")
    
    # Remove the incomplet words of the end.
    s = re.sub(ur'[^ ]*\u2026','', s)
    
    # Remove the line breacks.
    s = re.sub(r'\n', r' ', s)
    
    # Remove the URLs
    s = re.sub(r'https:\/\/t.co\/([a-z]|[0-9]){10}', r'', s)
    
    # Remove user names
    s = re.sub(r'@[^ ]*', r'', s)
    
    # Remove "rt"
    s = re.sub(r'rt', r'', s)
    
    s = re.sub(r'#', r'', s)
    
    # Replace successive spaces with a single space
    s = re.sub(r' +', r' ', s)
    
    # Remove all the characters that aren from ASCII code.
    s = re.sub(r'[^\x00-\x7f]',r'', s)
    
    print("Tweet filtrado: " + s + "\n")
    file.write(s + "\n")
```
